```python
from Windowing import *
from Plotting import Boxplots
```

# 0. Introduction
In this notebook we just see the effect of using the deprecated weights on MDS. Deprecated weights were given by $k^2$ for anchor points $s_k$. New weights are given by $2^k$

# 1. Load data

```python
win = Windowing(aggCurves)
win.set_points([16,   23,   32,   45,   64,   91,  128,  181,  256,  362,  512,
        724, 1024, 1448, 2048, 2896])

win.MDS()
MDS = win.results['MDS']['test error']

win.MDS(deprecated_weights=True)
MDS_deprecated = win.results['MDS']['test error']
```

```python
win.help()
```

# 2. Plot

```python
plot1 = {win.train_anchors[i]:MDS[:,:,i,-1] for i in range(len(win.train_anchors))}
plot2 = {win.train_anchors[i]:MDS_deprecated[:,:,i,-1] for i in range(len(win.train_anchors))}

Boxplots({'MDS':plot1, 'MDS deprecated':plot2}, ylim = [0,0.3], title='MDS with different weights', ylabel='Test error', xlabel='Anchor points', figsize=(3, 1), save = 'Plots/MDS_weights.png')
```
