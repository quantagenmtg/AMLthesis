```python
import matplotlib.pyplot as plt
import numpy as np
import torch

from Windowing import *
from Plotting import Compare, PlotHeatmap, HeatmapBoxPlot, Meanplots
```

```python
win = Windowing(aggCurves)
win.set_points([16,   23,   32,   45,   64,   91,  128,  181,  256,  362,  512,
        724, 1024, 1448, 2048, 2896, 4096])
win.MMF(steps = 500, lr = 0.1)
ErrorsHighLR = win.results['MMF']['train error']
```

```python
win.MMF(steps = 4000, lr = 0.01)
ErrorsLowLR = win.results['MMF']['train error']

win.MMF(steps = 8000, lr = 0.001)
ErrorsLowerLR = win.results['MMF']['train error']
```

```python
win.MMF(steps = 8000, lr = 0.0001)
ErrorsLowererLR = win.results['MMF']['train error']
```

```python
for lr in [0.1, 0.01, 0.001, 0.0001]:
    win.MMF(steps = 8000, lr = lr)
    Errors = win.results['MMF']['train error']
    plt.plot(np.nanmean(Errors, axis=(0,1)), label = str(lr) + ' LR')

plt.legend()
```

```python
# plot average error per window
plt.plot(np.nanmean(ErrorsHighLR, axis=(0,1)), label = '0.1 LR')

plt.plot(np.nanmean(ErrorsLowLR, axis=(0,1)), label = '0.01 LR')

plt.plot(np.nanmean(ErrorsLowerLR, axis=(0,1)), label = '0.001 LR')

plt.plot(np.nanmean(ErrorsLowererLR, axis=(0,1)), label = '0.0001 LR')

plt.legend()
```
